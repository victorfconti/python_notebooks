```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
aumentos_joao = np.array([90, 700, 2000, 5000])
aumentos_sarah = np.array([50, 450, 1500, 4500])
```

```python
plt.plot(aumentos_joao, c='Red', ls='--', marker='s', ms=4, label='João')
plt.plot(aumentos_sarah, c='Black', ls='--', marker='s', ms=4, label='Sarah')
plt.show()
```

```python
plt.plot(aumentos_joao, c='Red', ls='--', marker='s', ms=4, label='João')
plt.plot(aumentos_sarah, c='Black', ls='--', marker='s', ms=4, label='Sarah')
plt.legend()
plt.show()
```
